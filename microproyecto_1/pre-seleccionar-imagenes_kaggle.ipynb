```python
artist_list = [ 'frida kahlo',
 'vincent van gogh',
 'pierre auguste renoir',
 'claude monet',
 'rembrandt',
 'pablo picasso',
 'gustave dore',
 'henri matisse',
 'diego rivera',
 'salvador dali']
artist_list
```

```python
os_path = os.path.join('/kaggle/input/wikiart', 'classes.csv')

data_raw = pd.read_csv(os_path)
data_raw.head()
```

```python
data_artist = pd.DataFrame()
picture_list = []
for artist in artist_list:
    data_artist_pictures = data_raw[data_raw['artist'] == artist].sort_values('genre')
    count = data_artist_pictures.shape[0]
    index = count // 2
    picture = data_artist_pictures.iloc[index]
    picture_list.append(picture)
    
picture_df = pd.DataFrame(picture_list, columns=['filename', 'artist', 'genre']).sort_values('genre').reset_index(drop=True)
picture_df
```

```python
import shutil

input_dir = '/kaggle/input/wikiart'
output_dir = '/kaggle/working/pictures'

for index, row in picture_df.iterrows():
    src_path = os.path.join(input_dir, row['filename'])
    foldername, filename = os.path.split(src_path)
    dst_path = os.path.join(output_dir, filename)
    os.makedirs(os.path.dirname(dst_path), exist_ok=True)
    if os.path.exists(src_path):
        shutil.copy(src_path, dst_path)
        print(f"Copied {src_path} to {dst_path}")
    else:
        print(f"File not found: {src_path}")
```

```python
# Zip the contents of the output directory
shutil.make_archive('output_zip', 'zip', output_dir)
print("Zipped the contents of 'output' to 'output_zip.zip'")
```
